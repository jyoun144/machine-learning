### Step 0: Generate data for m=101 by n=2 matrix, where each column (n) represents a random variable (g1 & g2) and each row (m) represents all of the measurements/realizations of each random variable.

```python
import numpy as np
import matplotlib.pyplot as plt
g1 = np.arange(0,10+.1,.1)
g2 = g1**2
myMatrix =np.stack((g1, g2), axis=1)
print("Dimensions of original matrix:", myMatrix.shape)
plt.title("Original Matrix Measurements") 
plt.xlabel("g1 measurements")
plt.ylabel("g2 measurements")
plt.scatter(myMatrix[:,0], myMatrix[:,1], marker='.')
```

### Step 1:  Create mean-centered matrix from original data

```python
meanSubtractedMatrix = myMatrix - np.mean(myMatrix, axis=0)
print("Dimensions of mean-centered matrix:", myMatrix.shape)
plt.title("Mean-Centered Matrix Measurements") 
plt.xlabel("g1 measurements")
plt.ylabel("g2 measurements")
plt.scatter(meanSubtractedMatrix[:,0], meanSubtractedMatrix[:,1], marker='.')
```

### Step 2:  Create covariance matrix from mean-centered matrix

```python
expectedCovarianceMatrix = np.cov(meanSubtractedMatrix[:,0], meanSubtractedMatrix[:,1])
print("Expected Covariance Matrix:\n",expectedCovarianceMatrix)
numOfObservations = len(meanSubtractedMatrix[:,0])
var_g1 = sum((meanSubtractedMatrix[:,0] - np.mean(meanSubtractedMatrix[:,0]))**2)/(numOfObservations - 1)
var_g2 = sum((meanSubtractedMatrix[:,1] - np.mean(meanSubtractedMatrix[:,1]))**2)/(numOfObservations - 1)
cov_g1_g2 = sum((meanSubtractedMatrix[:,0] - np.mean(meanSubtractedMatrix[:,0]))*
                (meanSubtractedMatrix[:,1] - np.mean(meanSubtractedMatrix[:,1])))/(numOfObservations - 1)
covarianceMatrix = np.array([[var_g1, cov_g1_g2], [cov_g1_g2, var_g2]])
print("\nActual Covariance Matrix:\n",covarianceMatrix)
```

### Step 3:  Calculate eigenvectors and eigenvalues from covariance matrix

```python
eVals, eVecs = np.linalg.eig(covarianceMatrix)
maxMinIndex = np.argsort(-eVals)
eVals = eVals[maxMinIndex]
print("Eigenvalues in descending order:\n",eVals)
# Arrange eigen vectors in descending order according to the values of their associated eigenvalues
eVecs = np.stack((eVecs[:,maxMinIndex[0]], eVecs[:,maxMinIndex[1]]),  axis=1)
print("\nDimensions of eigenvectors matrix:\n", eVecs.shape)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
pcas = ['PCA 1', 'PCA 2']
ax.set_title('Proportion of Variance by PCA')
ax.bar(pcas,eVals/sum(eVals))
plt.show()
```

### Step 4: Generate scores matrix (Y = XP)

```python
scoresMatrix = meanSubtractedMatrix @ eVecs
print("Dimensions of scores matrix:\n", scoresMatrix.shape)
plt.title("Scores Matrix Data") 
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.scatter(scoresMatrix[:,0], scoresMatrix[:,1], marker='.')
```

### Step 5:  Partially recover data via PCA 1

```python
# Compute the outer product of two vectors
rotatedBackFromOneComponent = np.outer(scoresMatrix[:,0],  eVecs[:,0])
plt.title("PCA 1 Recovered Data") 
plt.scatter(rotatedBackFromOneComponent[:,0], rotatedBackFromOneComponent[:,1], marker='.', color="red", label='PCA 1')
plt.scatter(meanSubtractedMatrix[:,0], meanSubtractedMatrix[:,1], marker='.', label='mean-centered data')
plt.legend(loc="upper left")
```

### Step 6:  Fully recover data via PCA 1 and PCA 2

```python
rotatedBackFromBothComponents = scoresMatrix @eVecs
plt.title("PCA 1 & PCA 2 Recovered Data") 
plt.scatter(rotatedBackFromBothComponents[:,0], rotatedBackFromBothComponents[:,1], marker='o', color="red", label='PCA 1 & PCA 2')
plt.scatter(meanSubtractedMatrix[:,0], meanSubtractedMatrix[:,1], marker='.', label='mean-centered data')
plt.legend(loc="upper left")
```
